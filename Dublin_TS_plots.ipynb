```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import datetime as datetime
import matplotlib.patches as mpl_patches
%matplotlib inline
```

```python
df= pd.read_excel("dublin_data.xlsx", sheet_name='NO2')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel(("NO2 ($\mathregular{μg/m^{3}}$)"), fontsize=12)
ax.set_ylim(0, 80)

ax.plot(df['Date'], df['NO2_2020'], color='blue')   #marker = 'o'




#trend line

x = mdates.date2num(df['Date'])

y = df['NO2_2020'].values.flatten()

idx = np.isfinite(x)& np.isfinite(y)

z = np.polyfit(x[idx],y[idx],1)

p = np.poly1d(z)

ax.plot(x,p(x),"b--")

handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", 
                                 lw=0, alpha=0)] * 2
labels = []
labels.append("y=%.3fx+(%.3f)"%(z[0],z[1]))
#labels.append("r\u00b2 = (%.3f)"%(r_squared))


leg=ax.legend(handles, labels, loc='center', fontsize='8',
fancybox=True, framealpha=0.2, 
handlelength=0, handletextpad=0)
for text in leg.get_texts():
    plt.setp(text, color = 'blue')
        

plt.text(0.015, 0.64, 'WHO standard', transform=ax.transAxes , weight='bold', fontsize=8)
# the line equation:
print("y=%.6fx+(%.6f)"%(z[0],z[1]))

#ax.legend(loc='upper left', fontsize=14)
ax.tick_params(axis='y')
ax.axvline(x=df['Date'][26], color = 'red',linestyle="solid", lw=2)

#ax.axvspan(line1, line2, alpha=.8, color='green') fill b/w to lines
ax.axvline(x=df['Date'][41], color = 'grey',linestyle="dashed")
ax.axvline(x=df['Date'][66], color = 'grey',linestyle="dashed")
ax.axvline(x=df['Date'][79], color = 'grey',linestyle="dashed")
ax.axvline(x=df['Date'][96], color = 'grey',linestyle="dashed")
plt.axhline(y=50, color='green', linestyle='--')
plt.text(0.015, 0.64, 'WHO standard (24 hrs)', transform=ax.transAxes, color='green', weight='bold', fontsize=8)
plt.annotate("Pre-Lockdown", (0.09, 0.9), xycoords='axes fraction', fontsize=8)
plt.annotate("Phase-I", (0.32, 0.9), xycoords='axes fraction', fontsize=8)
plt.annotate("Phase-II", (0.51, 0.9), xycoords='axes fraction', fontsize=8)
plt.annotate("Phase-III", (0.68, 0.9), xycoords='axes fraction', fontsize=8)
plt.annotate("Phase-IV", (0.83, 0.9), xycoords='axes fraction', fontsize=8)
#rect = Rectangle(x=df['Date'][0:27], y =df['Date'][0:80],facecolor='lightgrey')
#plt.gca().add_patch(rect)


fig.tight_layout()  # otherwise the right y-label is slightly clipped

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
#plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
date_form = DateFormatter("%m-%d")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
# Puts x-axis labels on an angle
plt.gca().xaxis.set_tick_params(rotation = 90)
#plt.savefig('D:/AVINASH-IITM/Dublin/charts/NO2.png', dpi=500, bbox_inches='tight')
plt.show()
```
