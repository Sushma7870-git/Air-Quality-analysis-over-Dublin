# AOD Data Extraction

```python
import geemap
import ee
ee.Initialize()
import pandas as pd
```

```python
Map = geemap.Map(center = [53.35, -6.28], zoom=4)
Map
```

```python
roi = ee.Geometry.Point(-6.28, 53.35 ).buffer(2000)
Map.addLayer(roi)
Map
```

```python
AOD = ee.ImageCollection('MODIS/006/MCD19A2_GRANULES').filterDate('2020-05-01', '2020-05-31')
```

```python
def roi_mean(img):
    mean = img.reduceRegion(reducer=ee.Reducer.mean(), geometry=roi, scale=100).get('Optical_Depth_055')
    return img.set('date', img.date().format()).set('mean',mean)
```

```python
roi_reduced_imgs = AOD.map(roi_mean)
```

```python
imgs_list = roi_reduced_imgs.reduceColumns(ee.Reducer.toList(2), ['date','mean']).values().get(0)
```

```python
df = pd.DataFrame(imgs_list.getInfo(), columns=['date','mean'])
#df
```

```python
df['date'] = pd.to_datetime(df['date'])

df = df.set_index('date')
df
```
